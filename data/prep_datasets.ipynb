# All Datasets into one

```python
import pandas as pd

python_25k = pd.read_csv("python_25k.csv")
python_25k.head()
```

```python
python_500k = pd.read_csv("python_500k.csv")
python_500k.head()
```

```python
ml_questions = pd.read_csv("ml_questions.csv")
ml_questions.head()
```

```python
qa_python_ds_ml = pd.read_csv("qa_python_ds_ml.csv")
qa_python_ds_ml.head()
```

```python
print(python_25k.isna().sum())
print(python_500k.isna().sum())
print(ml_questions.isna().sum())
print(qa_python_ds_ml.isna().sum())
```

```python
python_500k = python_500k.dropna()
print(python_500k.isna().sum())
```

```python
prep_gemma = pd.concat([python_25k, python_500k, ml_questions, qa_python_ds_ml], ignore_index=True)
```

```python
prep_gemma = prep_gemma.sort_values(by="token_len_gemma", ascending=False)
prep_gemma.head()
```

```python
prep_gemma = prep_gemma[prep_gemma["text_len_gemma"] > 9]
prep_gemma = prep_gemma[prep_gemma["token_len_gemma"] < 257]
len(prep_gemma)
```

```python
prep_gemma.head()
```

```python
print(f"len: {len(prep_gemma)}")
prep_gemma = prep_gemma.drop_duplicates()
print(f"len without dups: {len(prep_gemma)}")
```

```python
prep_gemma = prep_gemma.drop(columns=["instruction", "output", "text_len_gemma", "token_len_gemma"])
```

```python
prep_gemma.head()
```

```python
prep_gemma.to_csv("gemma_331k.csv", index=False)
```
